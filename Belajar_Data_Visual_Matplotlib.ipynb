# Data Visual Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20)
y = x + 0.5

plt.plot(x, y)

plt.show()
```

```python
x1 = [1, 3, 5, 7, 9, 11]
y1 = [10, 25, 35, 33, 41, 50]

#plot datanya
plt.plot(x1,y1, label="Series-1", color="blue")

x2 = [2,4,6,8,10,12]
y2 = [15,29,32,33,38,55]

#plot data 2
plt.plot(x2,y2, label="Series-2", color="red")

#memberikan label untuk sumbu-x x-axis
plt.xlabel("Sumbu X")

#memberikan label untuk sumbu-y y-axis
plt.ylabel("Sumbu y")

plt.title("Multi Graph Line Python")

#tambahkan legend
plt.legend()

plt.show
```

```python
x = [3,5,7,9,11]
y = [10,25,35,33,59]

plt.scatter(x, y, alpha=0.5, c='red', marker='v')

plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Scatter Sederhana")
plt.show()
```

```python
#membuat sebuah data
subjects = [
    "Matematika", "Fisika", "Komunikasi", "Pemrograman Web"
]
score = [85, 62, 57, 95]
sub_new = []

for sub, sc in zip(subjects,score):
  sub_new.append(f"{sub}")

#memisahkan bagian
plt.pie(score,labels=subjects, explode=(0,0.1,0.1,0),
        startangle=90, #merotasi bagian
        shadow=True, #memberikan bayangan
        colors=["r", "g", "b", "y"], autopct = '%.1f%%')

# plt.pie(score, labels = subjects, colors=["r", "g", "b", "y"], autopct = '%.1f%%')
#%.1f% = menampilkan angka persentase

#judul
plt.title("Performa Developer Saat Ini")

plt.show()
```

```python
#membuat data baru
movie_rating = [1,2,3,4]
score = [21,34,67,35]

plt.bar(movie_rating, score, color="red")

plt.title("Movie Rating In Netflix")
plt.xlabel("Rating ")

plt.show()
```
