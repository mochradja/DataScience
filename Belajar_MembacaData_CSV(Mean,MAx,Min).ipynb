# Studi Kasus Data Event Organizer Indonesia

```python
df1 = pd.read_csv  ('https://gist.githubusercontent.com/yanwarsolah/4a5adbd418d92ef238fb45bf9d70205f/raw/27c231da2180579e99f0f8761d70d2ac05531226/data_event.csv')
df1
```

```python
#total budget event 3 selama beberapa tahun
e1_df = df1[df1.Event == "E3"]
print(e1_df)

print("Total budget :",e1_df['Budget'].sum())
```

```python
#cost budget tertinggi dari event 4 dalam beberapa tahun
e1_df = df1[df1.Event == "E4"]
print(e1_df)

print("Budget Tertinggi :",e1_df['Budget'].max())
```

```python
#pivot berdasarkan nama event sebagai index, tahun sebagai kolom, nilai sebagai budget
pivot3 = df1.pivot(index="Event", columns="Tahun", values="Income")
pivot3
```

```python
#income maksimum dari dalam beberapa tahun
e1_total = pivot3.loc['E1'].sum()
e2_total = pivot3.loc['E2'].sum()
e3_total = pivot3.loc['E3'].sum()
e4_total = pivot3.loc['E4'].sum()

print("E1", e1_total)
print("E2", e2_total)
print("E3", e3_total)
print("E4", e4_total)
income_max = max(e1_total, e2_total, e3_total, e4_total)
print("Income tertinggi :", income_max)
```

```python
#pivot berdasarkan nama event sebagai index, tahun sebagai kolom, nilai sebagai budget
pivot3 = df1.pivot(index="Event", columns="Tahun", values="Budget")
pivot3
```

```python
#nilai summary dari event 3 dari beberapa tahun
nilai_sum = pivot3.loc['E3'].sum()
print("nilai summary dari event 3 :", nilai_sum)
```

```python
#ambil max dan min untuk beberapa tahun dalam event
pivot3.loc[['E1', 'E2', 'E3', 'E4']].agg(['max', 'min'])
```

```python
#hitung nilai max dan min menggunakan aggregate event 3
pivot3.loc['E3'].agg(['max', 'min'])
```

```python
#kelompokkan data berdasarkan nama event lalu cari max dan min setiap tahunnya
pivot3.groupby('Event').agg(['max', 'min'])
```

```python
#dari pivot hitung nilai max dari total budget yang dikeluarkan setiap event
pivot3.loc['E1'].max
pivot3.loc['E2'].max
pivot3.loc['E3'].max
pivot3.loc['E4'].max
```
